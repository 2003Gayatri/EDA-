```python


# Assuming 'df' is your DataFrame containing the bank churn dataset

# Numerical attributes for outlier detection
numerical_attributes = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']

# Function to detect outliers using the IQR method
def detect_outliers_iqr(data_series):
    Q1 = data_series.quantile(0.25)
    Q3 = data_series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data_series[(data_series < lower_bound) | (data_series > upper_bound)]
    return outliers

# Loop through each numerical attribute and identify outliers
for attribute in numerical_attributes:
    outliers = detect_outliers_iqr(df[attribute])
    print(f"Outliers in {attribute}:")
    print(outliers)
```

```python


# Assuming 'df' is your DataFrame containing the bank churn dataset

# Function to handle outliers using the capping approach
def handle_outliers_capping(data_series, lower_bound, upper_bound):
    return data_series.clip(lower=lower_bound, upper=upper_bound)

# Handling outliers for the 'CreditScore' attribute
credit_score_lower_bound = 300
credit_score_upper_bound = 900
df['CreditScore'] = handle_outliers_capping(df['CreditScore'], credit_score_lower_bound, credit_score_upper_bound)

# Handling outliers for the 'Age' attribute
age_lower_bound = 18
age_upper_bound = 90
df['Age'] = handle_outliers_capping(df['Age'], age_lower_bound, age_upper_bound)

# Handling outliers for the 'NumOfProducts' attribute
num_of_products_lower_bound = 0
num_of_products_upper_bound = 4
df['NumOfProducts'] = handle_outliers_capping(df['NumOfProducts'], num_of_products_lower_bound, num_of_products_upper_bound)


```

### CAUSE AND EFFECT RELATIONSHIP ATTRIBUTES

```python
numerical_attributes = ['CreditScore', 'Age', 'Tenure', 'Balance',  'EstimatedSalary']
categorical_attributes = ['Geography', 'Gender', 'HasCrCard', 'IsActiveMember','NumOfProducts','Exited']

```

```python
 #Perform t-test for numerical attributes
numerical_significance = []

for attr in numerical_attributes:
    churned = df[df['Exited'] == 1][attr]
    not_churned = df[df['Exited'] == 0][attr]
    t_stat, p_value = ttest_ind(churned, not_churned)
    numerical_significance.append({'Attribute': attr, 'p-value': p_value})
t_test_results_df = pd.DataFrame(numerical_significance )
print(    numerical_significance)
```

```python
# Perform chi-square test for categorical attributes
categorical_significance = []
for attr in categorical_attributes:
    contingency_table = pd.crosstab(df[attr], df['Exited'])
    chi2_stat, p_value, _, _ = chi2_contingency(contingency_table)
    categorical_significance.append({'Attribute': attr, 'p-value': p_value})


chi2_results_df = pd.DataFrame(categorical_significance)
print(categorical_significance)
```

```python
print("Numerical Attributes:")
print(t_test_results_df)
print("\nCategorical Attributes:")
print(chi2_results_df)
```

```python

significance_results = pd.DataFrame(numerical_significance + categorical_significance)


significance_results = significance_results.sort_values(by='p-value')
print(significance_results )


#significant_attributes = significance_results[significance_results['p-value'] < 0.05]
#print(significant_attributes)
```

##Prediction Model

```python
# Identify the significant attributes
significant_attributes = [
    "Age",
    "Geography",
    "IsActiveMember",
    "Balance",
    "Gender",
    "NumOfProducts",
    "CreditScore",
]


# Encode the categorical attributes
encoder = LabelEncoder()
for attribute in significant_attributes:
    if df[attribute].dtype == "object":
        df[attribute] = encoder.fit_transform(df[attribute])



logistic_regression_model = LogisticRegression()
logistic_regression_model.fit(df[significant_attributes], df["Exited"])

# Create the decision tree model
decision_tree_model = DecisionTreeClassifier()
decision_tree_model.fit(df[significant_attributes], df["Exited"])

# Make predictions
encoder_filename = 'label_encoder.pkl'  # Specify the filename for the saved encoder
joblib.dump(encoder, encoder_filename)

logistic_regression_predictions = logistic_regression_model.predict(df[significant_attributes])
decision_tree_predictions = decision_tree_model.predict(df[significant_attributes])

# Evaluate the models
accuracy_logistic_regression = accuracy_score(df["Exited"], logistic_regression_predictions)
accuracy_decision_tree = accuracy_score(df["Exited"], decision_tree_predictions)

print("Accuracy of logistic regression:", accuracy_logistic_regression)
print("Accuracy of decision tree:", accuracy_decision_tree)
```

```python
plt.figure(figsize=(8, 6))
model_names = ['Logistic Regression', 'Decision Tree']
accuracies = [accuracy_logistic_regression, accuracy_decision_tree]
results_df = pd.DataFrame({'Model': model_names, 'Accuracy': accuracies})


sns.barplot(x='Model', y='Accuracy', data=results_df, palette='muted')
plt.ylim(0.0, 1.0)
plt.title('Accuracy of Logistic Regression and Decision Tree Models')
plt.ylabel('Accuracy')
plt.xlabel('Model')
plt.show()
```
